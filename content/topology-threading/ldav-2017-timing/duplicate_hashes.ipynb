```python
import numpy
import pandas
import toyplot.pdf

print("numpy version:   ", numpy.__version__)
print("pandas version:  ", pandas.__version__)
print("toyplot version: ", toyplot.__version__)
```

```python
#table_data = pandas.read_csv('hash_fight_collisions.csv')
table_data = pandas.read_csv('duplicate_hashes.csv')
#table_data
```

```python
pivot_data = table_data.pivot_table(values='collisions',
                                    index='num_cells',
                                    columns=['algorithm', 'topology'])
pivot_data
```

```python
algorithms = ['XOR', 'FNV1a', 'Morton']
topologies = ['regular', 'randomized']

canvas = toyplot.Canvas('4.5in', '2.7in')
axes = canvas.cartesian(bounds=(45,-15,10,-50),
                        xlabel='Number of Input Cells (Millions)',
                        ylabel='Hash Collisions (Billions)')

# Help toyplot to size the area large enough for the labels
axes.x.domain.max = 850
axes.x.ticks.locator = toyplot.locator.Explicit([0,200,400,600])

# This is a hack to change the default colors used by the plots.
# Hopefully future versions of toyplot will make setting this easier
palette = toyplot.color.brewer.palette('Paired')
import itertools
axes._palette = palette
axes._plot_colors = itertools.cycle(palette)
axes._scatterplot_colors = itertools.cycle(palette)
axes._text_colors = itertools.cycle(palette)

for alg in algorithms:
    for top in topologies:
        series = pivot_data[alg][top]
        x = series.index / 1000000.0
        y = numpy.array(series) / 1000000000.0
        axes.plot(x, y)

        if top == 'regular':
            name = alg + ' regular'
            baseline_shift = '50%'
        else: # top == 'randomized'
            name = alg + ' random'
            baseline_shift = '-50%'
        axes.text(x[-1], y[-1], name,
                  style={'text-anchor':'start',
                         '-toyplot-anchor-shift':'5pt',
                         'baseline-shift':baseline_shift})
```

```python
toyplot.pdf.render(canvas, 'duplicate_hashes.pdf')
```
